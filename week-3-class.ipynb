Consider an array $A=[a_1,a_2,\ldots,a_n]$ containing $n$ elements. The only operation you are allowed is to check whether two elements are equal, i.e. "is $a_i$ equal to $a_j$?" and these will be the elementary operations.

In what follows, assume that $n$ is a power of 2, i.e. $n=2^k$ for some positive integer $k$.

An element $x$ is the *majority* element in $A$ if it occurs more than $n/2$ times in the array. So for example, $[1,2,1,1]$ (with $n=4$) has a majority element 1 because it occurs $3>n/2$ times. The array $[1,2,3,3]$ has no majority element: 3 occurs twice, but this isn't enough because it is not true that $2>n/2$.

We will consider two distinct algorithms for finding the majority element in a list.

**(a)** Let us first start with a naive approach. Describe, in words, what the different steps of the following algorithm do and derive the complexity of this algorithm in terms of elementary operations. **[ 5 ]**
    
```Python
def majority(A):
    n = len(A) #O(1)
    for i in range(n): #O(n)
        c = 0
        for j in range(n):#O(n)
            if A[i]==A[j]:
                c = c+1
        if c>n/2:
            return A[i]
    return None
```

ANS: There are two for loops the first one picks the element to be compared against, where as the second one loops through the list again finding the frequency of the value that the first loop is currently pointing to.

If the frequency exceeds the necessary n/2 limit the value in the outer loop is returned. Due to the nested for loops which are each O(n) operations the overall complexity is O(n^2).

## Python template

````python
def majority(x):
    # should return an index i of the majority element
    # if there is one, so that x[i] is the majority element
    # or -1 if there is no majority element

    if len(x) == 1:
      return 0

    mid = len(x)//2
    L = x[:mid]
    R = x[mid:]
    m1 = majority(L)
    m2 = majority(R)

    if m1 != -1:
      if mid<x.count(x[m1]):
        return m1
      else:
        m1 = -1

    if m2 != -1:
      m2 += mid
      if mid<x.count(x[m2]):
        return m2
      else:
        m2 = -1

    if m1 == -1 and m2 == -1:
      return -1



x1 = [1,2,3,4,3,2,2,2,2,2,2,3,2,1,2,3]
x2 = [8,9,10,10]
maj1 = majority(x1)
maj2 = majority(x2)
print(maj1, maj2, x1[maj1])
if maj1==-1:
    print("Wrong! (1)")
elif maj1>=len(x1):
    print("Wrong! (2)")
elif x1[maj1]!=2:
    print("Wrong! (3)")
elif maj2!=-1:
    print("Wrong! (4)")
else:
    print("You got it!")
````

C++ Text

````python
#include<iostream>
#include<vector>
#include<algorithm>

using namespace std;

int majority(vector<int> &x)
{
    // should return i where either i==-1 if there is no majority
    // element, or if 0<=i<x.size() where x[i] is the majority
    // element.

    if(x.size()==1) return 0;

    int mid = x.size()/2;

    vector<int> L;
    vector<int> R;

    L.assign(x.begin(), x.begin()+mid);
    R.assign(x.begin()+mid, x.end());

    int m1 = majority(L);
    int m2 = majority(R);

    if(m1 != -1){
        if(mid<count(x.begin(), x.end(), x[m1])){
            return m1;
        } else {
            m1 = -1;
        }
    }

    if(m2 != -1){
        m2 = m2 + mid;
        if(mid<count(x.begin(), x.end(), x[m2])){
            return m2;
        } else {
            m2 = -1;
        }
    }

    return -1;

}

void WRONG(char x){
    cout<< "Wrong! " << x <<endl;
}

int main(void)
{
    vector<int> x1{1,2,3,4,3,2,2,2,2,2,2,3,2,1,2,3};
    vector<int> x2{8,9,10,10};
    int maj1 = majority(x1);
    int maj2 = majority(x2);
    //cout << maj1 << " " << maj2 << endl;
    if(maj1==-1) WRONG('1');
    if(maj1>=x1.size()) WRONG('2');
    if(x1[maj1]!=2) WRONG('3');
    if(maj2!=-1) WRONG('4');
    cout << "You got it!" << endl;
    return 0;
}
````
