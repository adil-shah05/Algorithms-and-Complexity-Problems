The advertising company "Greedy Oligarchs & Other Grubby Lowdown Evildoers" has $n$ clients and wants to maximise their revenues before their initial public offering in $T$ days. They decide to concentrate each day on a single advertising campaign for a single client. For example, on day 1 they might advertise for client 3, on day 2 for client 7 and day 3 back to client 3 again. Some clients are bigger than others and so make more money, but repeatedly advertising for the same clients has diminishing returns. The first day of advertising for client $i$ they will make $c_i$. The second day of advertising for the same client, $c_i/2$, the third day $c_i/4$ and so on, at day $k$ for client $i$ they will make $c_i/2^k$, so it might make sense for them to advertise one day with one client, and then another day with another client, and so on.

3. What is the complexity of your algorithm? Note you should give your answer in terms of $n$ and $T$. (This two variable complexity notation is not examinable.)

#First Try

Does a for loop through the number of days sorting the array each time so that the most profitable is always at the end of the array. This end value is added to the total then amended to half of it before restarting the loop.

Since we know after sorting the most profitable choice is at the end the greedy choice algorithm holds. Time complexity is O(n log(n)) to sort and O(T) to go through the days. therefore there is an overall complexity of O(T * n log(n))

````python
clients = [2,3,5] #n = 3

T = 10 #Ten days
ci = 0
sum = 0

for n in range(T):
  clients.sort()
  ci = clients[-1]
  sum += ci
  ci = ci/2
  clients[-1] = ci

print(sum)

````

````python
class MaxFibonacciHeap:
    def __init__(self):
        self.min_heap = FibonacciHeap()  # underlying min-heap

    def insert(self, value, client_id, times_used):
        # Insert NEGATIVE value
        self.min_heap.insert((-value, client_id, times_used))

    def extract_max(self):
        neg_value, client_id, times_used = self.min_heap.extract_min()
        # Negate back to get original value
        return (-neg_value, client_id, times_used)

    def is_empty(self):
        return self.min_heap.is_empty()


def greedy(T, heap):
  ci = 0
  sum = 0

  for n in range(T):
    ci = heap.extract_max()
    sum += ci
    heap.insert(ci/2)

  print(sum)

clients = [2, 3, 5]
heap = MaxFibonacciHeap()

for i in clients:
  heap.insert(i)

greedy(10, heap)
````
